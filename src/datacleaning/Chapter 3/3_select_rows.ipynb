# Select the last three rows using tail and [] operator slicing

```python
nls97.tail(3).T
```

```python
nls97[-3:].T
```

```python
# Every row except last 3 rows
# By not providing a value to the right of the colon in [-3:],
# we are telling the accessor to get all rows from the third-to-last row to the end of the DataFrame

nls97[:-3].T
```

# Select a few rows using the loc data accessor

```python
# Note that nls97.loc[[195884,195891,195970]] and nls97.loc[195884:195970] return the same DataFrame

nls97.loc[[195884, 195891, 195970]].T
```

# Select multiple rows conditionally using boolean indexing

```python
nls97['nightlyhrssleep'].quantile(0.05)
```

```python
nls97['nightlyhrssleep'].count()
```

```python
sleepcheckbool = nls97['nightlyhrssleep'] <= 4
```

```python
sleepcheckbool
```

```python
# nls97.loc[nls97['nightlyhrssleep'] <=4]
lowsleep = nls97.loc[sleepcheckbool]
```

```python
nls97.loc[nls97['nightlyhrssleep'] <= 4]
```

```python
lowsleep.shape
```

# Select rows based on multiple conditions

```python
# Similar to lowsleep DataFrame – lowsleep3pluschildren = lowsleep.loc[lowsleep.childathome>=3]

lowsleep['childathome'].describe()
```

```python
lowsleep3pluschildren = nls97.loc[(nls97['nightlyhrssleep'] <= 4)
                                  & (nls97['childathome'] >= 3)]
```

```python
lowsleep3pluschildren.shape
```

# Select rows and columns based on multiple conditions

```python
lowsleep3pluschildren = nls97.loc[(nls97['nightlyhrssleep'] <= 4) &
                                  (nls97['childathome'] >= 3),
                                  ['nightlyhrssleep', 'childathome']]
```

```python
lowsleep3pluschildren
```
