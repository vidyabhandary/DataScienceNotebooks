```python
# xlrd 2.0.1 does not support xlsx files. Upgraded pandas to 1.2 and installed openpyxl. It automatically uses that 
# openpyxl engine

percapitaGDP = pd.read_excel('data/GDPpercapita.xlsx', 
                             #engine='openpyxl',
                             sheet_name='OECD.Stat export',
                             skiprows=4,
                             skipfooter=1,
                             usecols='A, C:T')
```

```python
percapitaGDP.head()
```

```python
percapitaGDP.info()
```

```python
percapitaGDP.rename(columns={'Year' : 'metro'}, inplace=True)
```

```python
# Check for leading spaces
percapitaGDP['metro'].str.startswith(' ').any()
```

```python
# Check for trailing spaces
percapitaGDP['metro'].str.endswith(' ').any()
```

```python
# Remove the starting and ending spaces
percapitaGDP['metro'] = percapitaGDP['metro'].str.strip()
```

```python
# Iterate over all of the GDP year columns (2001-2018) and convert the data type
# from object to float. Coerce the conversion even when there is character data

for col in percapitaGDP.columns[1:]:
    percapitaGDP[col] = pd.to_numeric(percapitaGDP[col], errors='coerce')
    percapitaGDP.rename(columns={col:'pcGDP'+col}, inplace=True)
```
